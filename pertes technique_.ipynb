```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/Users/abderrahim_boussyf/Desktop/pertes technique.xlsx'  # Replace with your file path
sheet_name = 'Sheet1'  # Replace with your sheet name if different

# Load the data
df = pd.read_excel(file_path, sheet_name=sheet_name)


# Renommer les colonnes en fonction de celles que vous avez
df.columns = [
    'Index', 'Direction Provinciale',
    'Achat Energie avant transformateurs postes sources annuel (MWhrs)',
    'Perte technique HT-MT', 'Colonne_4', 'Perte technique Rseau MT',
    'Colonne_6', 'Ventes MT annuelles', 'Achats BT annuels',
    'Perte technique Rseau BT', 'Colonne_10', 'Perte technique Branchement BT',
    'Colonne_12', 'Perte technique Compteurs', 'Colonne_14',
    'Perte technique global', 'Taux de perte global'
]

# Conversion des colonnes en numriques
cols_to_convert = [
    'Achat Energie avant transformateurs postes sources annuel (MWhrs)',
    'Perte technique HT-MT', 'Perte technique Rseau MT',
    'Ventes MT annuelles', 'Achats BT annuels',
    'Perte technique Rseau BT', 'Perte technique Branchement BT',
    'Perte technique Compteurs', 'Perte technique global'
]

for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Calcul des pertes techniques
df['Pertes_HT_MT'] = (df['Perte technique HT-MT'] / df['Achat Energie avant transformateurs postes sources annuel (MWhrs)']) * 100
df['Pertes_Rseau_MT'] = (df['Perte technique Rseau MT'] / df['Ventes MT annuelles']) * 100
df['Pertes_Rseau_BT'] = (df['Perte technique Rseau BT'] / df['Achats BT annuels']) * 100
df['Pertes_Branchement_BT'] = (df['Perte technique Branchement BT'] / df['Achats BT annuels']) * 100
df['Pertes_Compteurs'] = (df['Perte technique Compteurs'] / df['Ventes MT annuelles']) * 100
df['Pertes_Globale'] = (df['Perte technique global'] / df['Achat Energie avant transformateurs postes sources annuel (MWhrs)']) * 100

# Visualisation des pertes
plt.figure(figsize=(14, 10))

# Graphique des pertes techniques par catgorie
plt.subplot(2, 1, 1)
df[['Pertes_HT_MT', 'Pertes_Rseau_MT', 'Pertes_Rseau_BT', 'Pertes_Branchement_BT', 'Pertes_Compteurs']].mean().plot(kind='bar', ax=plt.gca())
plt.title('Pertes Techniques Moyennes par Catgorie')
plt.ylabel('Pourcentage (%)')

# Graphique des pertes techniques globales
plt.subplot(2, 1, 2)
plt.plot(df['Pertes_Globale'].dropna(), marker='o', label='Pertes Globales')
plt.title('Pertes Techniques Globales')
plt.ylabel('Pourcentage (%)')
plt.xlabel('Index')

plt.tight_layout()
plt.show()
```
