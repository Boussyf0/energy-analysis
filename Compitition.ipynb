```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# One-hot encoding categorical columns using sparse format
ohe = OneHotEncoder(sparse_output=True)

# List of categorical columns
categorical_cols = ['Gender', 'Marital Status', 'Education Level', 'Occupation', 
                    'Location', 'Policy Type', 'Smoking Status', 'Exercise Frequency', 
                    'Property Type', 'Policy Start Date', 'Customer Feedback']

# One-hot encoding directly
encoded_data_sparse = ohe.fit_transform(df[categorical_cols])

# Convert sparse matrix to a DataFrame
encoded_df = pd.DataFrame.sparse.from_spmatrix(
    encoded_data_sparse,
    columns=ohe.get_feature_names_out(categorical_cols)
)

# Concatenate the sparse DataFrame with the original DataFrame
df = pd.concat([df.reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)
```

If the code takes too much time, it might be due to the size of the dataset or the high number of categorical features being processed for one-hot encoding. Here are optimization techniques to reduce execution time
1. Use sparse=True for OneHotEncoder
By default, OneHotEncoder outputs a sparse matrix if (sparse_output=True). You can skip the .toarray() step, which converts it to a dense array and consumes memory.
2. Reduce the Number of Categorical Features
Processing a large number of categorical columns, especially high-cardinality ones, can slow things down. Evaluate and exclude unnecessary columns or preprocess some features differently (e.g., label encoding for ordinal columns).
3. Parallel Processing
One-hot encoding can be parallelized using libraries like joblib or dask. For example, you can process categorical columns in chunks.
