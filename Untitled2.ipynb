```python
import pandas as pd
import matplotlib.pyplot as plt

# Définir les tâches du projet et leur durée (en jours)
tasks = {
    'Analyse des besoins': 5,
    'Conception technique': 4,
    'Configuration environnement': 3,
    'Développement du pipeline ETL': 7,
    'Tests unitaires': 3,
    'Déploiement': 4,
    'Documentation': 3,
    'Finalisation et revue': 2,
}

# Dates de début des tâches
start_dates = pd.date_range('2024-08-01', periods=len(tasks), freq='7D')

# Créer un DataFrame pour les tâches
gantt_data = pd.DataFrame({
    'Task': list(tasks.keys()),
    'Start Date': start_dates,
    'Duration': list(tasks.values())
})

# Calculer les dates de fin
gantt_data['End Date'] = gantt_data['Start Date'] + pd.to_timedelta(gantt_data['Duration'], unit='D')

# Création du diagramme de Gantt
plt.figure(figsize=(10, 6))
plt.barh(gantt_data['Task'], gantt_data['Duration'], left=gantt_data['Start Date'].apply(lambda x: x.toordinal()), color='skyblue')
plt.xlabel('Dates')
plt.ylabel('Tâches')
plt.title('Diagramme de Gantt du projet')
plt.grid(True)

# Personnaliser l'axe des x avec les dates
ax = plt.gca()
ax.set_xticks([date.toordinal() for date in gantt_data['Start Date']])
ax.set_xticklabels([date.strftime('%Y-%m-%d') for date in gantt_data['Start Date']])

plt.tight_layout()
plt.show()
```
